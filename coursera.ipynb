```python
import pandas as pd
from bs4 import BeautifulSoup

# Parse HTML with BeautifulSoup
soup = BeautifulSoup(html_data_2, "html.parser")

# Find all tables in the HTML
tables = soup.find_all("table")

# Convert the first table to a DataFrame
gme_revenue = pd.read_html(str(tables[0]))[0]

# Inspect the first few rows to see what the columns are
print(gme_revenue.head())

# Rename columns to 'Date' and 'Revenue' (adjust based on the inspection)
gme_revenue.columns = ['Date', 'Revenue']

# Remove $ and commas from Revenue column
gme_revenue['Revenue'] = gme_revenue['Revenue'].str.replace('$', '', regex=True)
gme_revenue['Revenue'] = gme_revenue['Revenue'].str.replace(',', '', regex=True)

# Convert Revenue to numeric
gme_revenue['Revenue'] = pd.to_numeric(gme_revenue['Revenue'], errors='coerce')

# Display the cleaned DataFrame
print(gme_revenue.head())
```

```python
import pandas as pd
from bs4 import BeautifulSoup

# Parse HTML with BeautifulSoup
soup = BeautifulSoup(html_data_2, "html.parser")

# Find all tables in the HTML
tables = soup.find_all("table")

# Convert the first table to a DataFrame
gme_revenue = pd.read_html(str(tables[0]))[0]

# Rename columns
gme_revenue.columns = ['Date', 'Revenue']

# Remove $ and commas, strip whitespace
gme_revenue['Revenue'] = gme_revenue['Revenue'].replace({'\$': '', ',': ''}, regex=True).str.strip()

# Drop rows where Revenue is empty or null
gme_revenue = gme_revenue[gme_revenue['Revenue'] != ""]
gme_revenue = gme_revenue.dropna(subset=['Revenue'])

# Convert Revenue to numeric (now safe)
gme_revenue['Revenue'] = pd.to_numeric(gme_revenue['Revenue'])

# Display the cleaned DataFrame
print(gme_revenue.head())
```

```python
import matplotlib.pyplot as plt

# Example make_graph function
def make_graph(stock_data, title):
    plt.figure(figsize=(12,6))
    plt.plot(stock_data['Date'], stock_data['Close'], color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# Filter Tesla data up to June 2021
tesla_data['Date'] = pd.to_datetime(tesla_data['Date'])
tesla_plot_data = tesla_data[tesla_data['Date'] <= '2021-06-30']

# Plot the graph
make_graph(tesla_plot_data, "Tesla Stock Price Up to June 2021")
```

```python
# Ensure the Date columns are in datetime format
gme_data['Date'] = pd.to_datetime(gme_data['Date'])
gme_revenue['Date'] = pd.to_datetime(gme_revenue['Date'])

# Filter both stock and revenue data up to June 2021
gme_stock_filtered = gme_data[gme_data['Date'] <= '2021-06-30']
gme_revenue_filtered = gme_revenue[gme_revenue['Date'] <= '2021-06-30']

# Call make_graph with only the two arguments it expects
make_graph(gme_stock_filtered, gme_revenue_filtered)
```
